## **THEORY**

#### **1. Conditional Probability :**  

Conditional Probability of event $E$ given that event $F$ has already occured is defined as : 
$$ P(E|F) = \frac{Number\space of\space elementary\space events\space favourable\space to\space E\cap F}{Number\space of\space elementary\space events\space favourable\space to\space F}  
= \frac{n(E\cap F)}{n(F)}  
= \Large\frac{\frac{n(E\cap F)}{n(S)}}{\frac{n(F)}{n(S)}}
\normalsize= \frac{P(E\cap F)}{P(F)} $$

Hence,

$$ P(E|F) P(F) = P(E\cap F) $$  
$$ and $$  
$$ P(F|E) P(E) = P(F\cap E) = P(E\cap F) $$  

$$\Large\therefore\space \normalsize P(E|F) = \frac{P(F|E)\space P(E)}{P(F)} $$

---
#### **2. Theorem of Total Probability :**  

Let $ {E_{1}, E_{2},...,E_{n}} $ be a partition of the sample space S, and suppose that each of the events $ E_{1}, E_{2},...,E_{n} $ has nonzero probability of occurrence.  
Let $A$ be any event associated with $S$, then  
$$ P(A) = P(E_{1})\space P(A|E_{1}) + P(E_{2})\space P(A|E_{2}) + ... + P(E_{n})\space P(A|E_{n})  
= \Large\sum_{j=1}^{n} \normalsize P(E_{j}) P(A|E_{j}) $$

---
#### **3. Bayes’ Theorem :**  

According to **Bayes' Theorem**, if $ E_{1}, E_{2},...,E_{n} $ are n non empty events which constitute a partition of sample space $S$, i.e. $ E_{1}, E_{2},...,E_{n} $ are pairwise disjoint and $ E_{1}\cup E_{2}\cup...\cup E_{n} = S $ and $A$ is any event of nonzero probability, then
$$ P(E|F) = \frac{P(E∩F)}{P(F)}  
= \frac{P(F|E)\space P(E)}{P(F)}  
= \frac{P(F|E)\space P(E)}{\Large\sum_{j=1}^{n} \normalsize P(E_{j})\space P(F|E_{j})} \;\;\;\;\;\; \normalsize\forall i = 1, 2, 3, ..., n $$  

**Bayes' Theorem in the language of Machine Learning :** 

In machine learning we are often interested in determining the best hypothesis from some space $H$, given the observed training data $D$.  
One way to specify, the concept of best hypothesis is to say that we demand the most probable hypothesis, given the data D plus any initial knowledge about the prior probabilities of the various hypotheses in $H$.  

_Bayes' theorem provides a way to calculate the probability of a hypothesis based on its prior probability, the probabilities of observing various data given the hypothesis, and the observed data itself._
$$ P(h|D) = \frac{P(h|D)\space P(h)}{P(D)} $$

1. $ P(h|D) $ is called the **posterior probability of** $h$, because it reflects our confidence that $h$ holds after we have seen the training data $D$.
2. $ P(D|h) $ is often referred to as **likelihood of the data** $D$ given $h$.
3. $ P(h) $ is often called the **prior probability of** $h$ and may reflect any background knowledge we have about the chance that $h$ is a correct hypothesis.
4. $ P(D) $ is often called the **marginal probability of** $D$.

---
#### **4. Bayes' Optimal Classifier :** 

In general, the most probable classification of the new instance is obtained by combining the predictions of all hypotheses, weighted by their posterior probabilities.
If the possible classification of the new example can take on any value $v_{j}$ from some set $V$, then the probability $P(v_{j}|D)$ that the correct classification for the new instance is $v_{j}$, is just  

$$ P(v_{j}|D) = \Large\sum_{h_{i}\in H} \normalsize P(v_{j}|h_{i})\space P(h_{i}|D) $$

The optimal classification of the new instance is the value $v_{j}$, for which $P(v_{j}|D)$ is maximum.

**Bayes' Optimal Classification**

$$ \max_{v_{j} \in V} \Large\sum_{h_{i}\in H} \normalsize P(v_{j}|h_{i})\space P(h_{i}|D) $$  

* Any system that classifies new instances according to above equation is called a Bayes optimal classifier, or Bayes optimal learner. **No other classification method using the same hypothesis space and same prior knowledge can outperform this method on average.**

* This method maximizes the probability that the new instance is classified correctly, given the available data, hypothesis space, and prior probabilities over the hypotheses.  

* However in reality, it is quite difficult to pre-determine the posterior probabilities for any given hypotheses space $H$ and dataset $D$. Hence, Bayes' Optimal Classifier is not feasible to use in real world scenarios.

---
#### **5. Naive Bayes' Classifier :**  

**_A Naive Bayes' Classifier assumes that the presence of a particular feature in a class is unrelated to the presence of any other feature._**

* In many learning scenarios, the learner considers some set of candidate hypotheses $H$ and is interested in finding the most probable hypothesis $h \in H $ given the observed data $D$ (or at least one of the maximally probable if there are several). Any such maximally probable hypothesis is called a **maximum a posteriori (MAP) hypothesis**.  

* We can determine the **MAP hypothesis** by using **Bayes' theorem** to calculate the posterior probability of each candidate hypothesis. More precisely, we will say that $ h_{MAP}$ is a _MAP hypothesis_ provided  

$$ h_{MAP} = \max_{h \in H} \normalsize P(h|D)  
= \max_{h \in H} \normalsize\frac{P(h|D)\space P(h)}{P(D)}  
= \max_{h \in H} \normalsize P(h|D)\space P(h) $$

> **Notice in the final step above we dropped the term $P(D)$ because it is a constant independent of $h$.**  

* As we know, $P(D|h)$ is called the **likelihood of the data** $D$ given $h$, and any hypothesis that maximizes $P(D|h)$ is called a **maximum likelihood (ML)** hypothesis,  

$$ h_{ML} = \max_{h \in H} \normalsize P(D|h) $$  
* The fundamental Naive Bayes' assumption is that each feature makes an:
    - **independent** &
    - **equal**  
contribution to the outcome. Therefore,  
$$ h_{ML} = \max_{h \in H} \normalsize P(D|h)  
= \max_{h \in H} \Large\prod_{i=1}^{n} \normalsize P(d_{i}|h) $$
$$ where\space n = no.\space of\space features\space in\space the\space data\space instance\space D $$  
* The Naive Bayes' classifier applies to learning tasks where each instance $x$ is described by a conjunction of attribute values and where the target function $f(x)$ can take on any value from some finite set $V$. 
* A set of training examples of the target function is provided, and a new instance is presented, described by the tuple of attribute values $[a_{1}, a_{2},...a_{n}]$. The learner is asked to predict the target value, or classification, for this new instance.  
* The Bayesian approach to classifying the new instance is to assign the most probable target value, $v_{MAP}$ given the attribute values $[a_{1}, a_{2},...a_{n}]$ that describe the instance.
$$ v_{MAP} = \max_{v_{j}\in V} P(v_{j}|[a_{1}, a_{2},...a_{n}])$$  
* We can use Bayes' theorem to rewrite this expression as  
$$ v_{MAP} = \max_{v_{j}\in V} \frac{P([a_{1}, a_{2},...a_{n}]|v_{j})\space P(v_{j})}{P([a_{1}, a_{2},...a_{n}])}  
= \max_{v_{j}\in V} P([a_{1}, a_{2},...a_{n}]|v_{j})\space P(v_{j})$$  
&emsp;&emsp; Since, the event for each feature $a_{i}$ to be some $x_{i}$ is independent of each other, then, according to the theorem of Total Probability,  
$$ P([a_{1}, a_{2},...a_{n}]) = \Large\sum_{j=1}^{m} \normalsize P(v_{j}) P([a_{1}, a_{2},...a_{n}]|v_{j})  
= \Large\sum_{j=1}^{m}( \normalsize P(v_{j}) \Large\prod_{i=1}^{n} \normalsize P(a_{i}|v_{j}) \Large) $$  
$$where\space m = no.\space of\space distinct\space classes\space in\space the\space outcome$$

**Naive Bayes' Classification**  
$$ \max_{v_{j}\in V}  P(v_{j}) \Large\prod_{i=1}^{n} \normalsize P(a_{i}|v_{j}) $$
$  \Large\therefore \;\;\normalsize The\; actual\; conditional\; probability\; of\; the\; target\; value\; is\; given\; as-\; $
$$ \max_{v_{j}\in V}  \frac{P(v_{j}) \Large\prod_{i=1}^{n} \normalsize P(a_{i}|v_{j})}{\Large\sum_{j=1}^{m}( \normalsize P(v_{j}) \Large\prod_{i=1}^{n} \normalsize P(a_{i}|v_{j}) \Large)}$$  

---
#### **6. Estimating Probabilities :**  

If the training dataset is such that one of the conditional probabilities is zero, then the entire expression for that particular outcome becomes zero. In order to avoid such scenarios we can adopt a Bayesian approach to estimate the probability, using the m-estimate defined as follows-  
$$ m-estimate: P(x_{i} = c|v_{j}) = \frac{n_{c} + m\space p}{n + m} $$
$$ where\space n_{c} = no.\space of\space times\space the\space event\space is\space observed\space to\space occur $$
$$ n = total\; no.\; of\; oppurtunities $$
$$ p = \frac{1}{no.\; of\; possible\; values\; of\; the\; attribute\; x_{i}} $$
$$ m = equivalent\; sample\; size\; (for\; laplace\; smoothing\; m = no.\; of\; possible\; values\; of\; the\; attribute\; x_{i}) $$

## **IMPLEMENTATION**

### **1. Importing the Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

### **2. Importing the Database**

```python
filePath = input("Enter absolute file path of the dataset: ")
print('The file path is:', filePath)
data = pd.read_csv(filePath)
print('The data is as follows:')
data
```

### **3. Separating Independent and Dependent Variables**

```python
x = data.iloc[:, :-1].values
print('Independent Variables :')
print(x)

y = data.iloc[:, -1].values
print('Dependent Variables :')
print(y)
```

### **4. Splitting the Dataset into Training and Test Set**  

**Implementation :**
  * LIBRARY: **sklearn**
  * MODULE: **model_selection**
  * FUNCTION : test_train_split() :- **Split arrays or matrices into random train and test subsets.**

**Parameters :**

1. **arrays : sequence of indexables with same length / shape[0]:** 

  Allowed inputs are lists, numpy arrays, scipy sparse matrices or pandas dataframes.

2. **test_size: float or int, default=None**

  * If **float**, should be between 0.0 and 1.0 and represent the proportion of the dataset to include in the test split. 
  * If **int**, represents the absolute number of test samples.
  * If **None**, the value is set to the complement of the train size. 
  * If train_size is also None, it will be set to 0.25.

3. **train_size: float or int, default=None**

  * If float, should be between 0.0 and 1.0 and represent the proportion of the dataset to include in the train split. 
  * If int, represents the absolute number of train samples.  
  * If None, the value is automatically set to the complement of the test size.

4. **random_state : int, RandomState instance or None, default=None**

  Controls the shuffling applied to the data before applying the split. Pass an int for reproducible output across multiple function calls.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)

print('X → TRAIN :')
display(x_train)

print('Y → TRAIN :')
display(y_train)

print('X → TEST :')
display(x_test)

print('Y → TEST :')
display(y_test)
```

### **5. Feature Scaling**  

* **What is Feature Scaling?** 

> Feature Scaling is putting all features in same scale.  

* **Why Feature Scaling?**

>Some models of Machine Learning requires the features to be in same scale. If they are not in same scale the feature with much larger scale will dominate the feature with smaller values and therefore the accuracy of the ML model will be compromised.

* **Feature Scaling Techniques :**

    **1. Standardisation:** puts all the values of the feature in range of [-3, 3]

  $$ x_{stand} = \frac{x - mean(x)}{standard\; deviation(x)} $$  
                            
    **2. Normalisation:** puts all the values of the feature in range of [0, 1]

  $$ x_{norm} = \frac{x- min(x)}{max(x) - min(x)} $$

    **_Standardisation is recommended as it works in almost all cases._**  

* **Implementation :**  

  * LIBRARY : **sklearn**
  * MODULE : **preprocessing**
  * CLASS : **StandardScalar** _or_ **Normalizer**

* **Steps :**  
  * Create an object of the class.
  * Call fit_transform method on the training data set where ‘fit’ will find the mean and standard deviation of values and ‘transform’ method will actually use the standardisation formula on the x values to complete their scaling.
  * Repeat for Test set except the fit method will not be called for test set, **the mean and standard deviation of Training set will be used therefore only transform function will be called**.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

print('X → TRAIN :')
display(x_train)

print('X → TEST :')
display(x_test)
```

### **6. Training the Naive Bayes' Classification Model**  

**Implementation :**
  * LIBRARY : **sklearn**
  * MODULE : **naive_bayes**
  * CLASS : **GaussianNB**  
  _Other Classes included in the module are **BernoulliNB**, **MultinomialNB**, **CategoricalNB** etc._

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

### **7. Predicting Test Set Results**

```python
y_pred = np.array(classifier.predict(x_test))

print('Y → PRED :')
display(y_pred)
```

### **8. Visualizing Test vs. Predicted Results**

```python
fig1 = plt.figure(figsize=[1,3], dpi=100)

plot1 = fig1.add_subplot(1, 1, 1)
testVsPred = pd.DataFrame({'Y → TEST':y_test, 'Y → PRED':y_pred})
display(testVsPred)
testVsPred.plot(marker='o', ax=plot1)

plot1.set_xlabel('X → TEST record serial number', fontsize = 12)
plot1.set_ylabel('Label Value', fontsize = 12)
plot1.set_title('Y → TEST vs Y → PRED', fontsize = 17)
plot1.legend()

fig1.subplots_adjust(left=1, right=15, top=2.5, wspace=0.15, hspace=0.2)
```

### **9. Confusion Matrix and Accuracy Score**

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print("The Confusion Matrix is :")
print(cm)

accuracy =  accuracy_score(y_test, y_pred)
print("The accuracy of the model is = {a} %".format(a = accuracy * 100))
```
